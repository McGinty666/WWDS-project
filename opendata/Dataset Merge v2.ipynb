```python
# Need to make sure columns names are correct otherwise xml won't save (have to use "LSOA" not "LSOA or Data zone")
# Need to align all of the column headers
# Need to make sure that the file names are what I want to populate the Data Provider column with

# Function for generating and replacing Sample_Id with UUIDs
def generate_uuid_for_unique_entries(column):
    unique_values = column.unique()
    uuid_dict = {value: str(uuid.uuid4()) for value in unique_values}
    return column.map(uuid_dict)


# Creating a list of all csv's in Data dir
a = [x for x in os.listdir('Data/') if x[-4:]=='.csv']


for dataset in a[-2:-1]:
    print(dataset)
    
    with open(f'Data/{dataset}', 'rb') as rawdata:
        result = chardet.detect(rawdata.read(10000))
    print(result)
    
    
    df = pd.read_csv(f'Data/{dataset}', encoding=result['encoding'].lower())
    df['Data_Provider'] = dataset[:-4]
    df['Sample_Id'] = generate_uuid_for_unique_entries(df['Sample_Id'])
    display(df)
    
    df_csv = df
    df_xml = df
    df_json = df
    
    
    df_csv.to_csv(f'Data/Ready/csv/{dataset[-4:]}.csv')
    df_xml.to_xml(f'Data/Ready/xml/{dataset[-4:]}.xml')
    df_json.to_json(f'Data/Ready/json/{dataset[-4:]}.json')
```
